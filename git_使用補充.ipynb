# git
0. [建立自己的 git 帳號](https://github.com)
1. [download git](https://git-scm.com/download/win)
2. 安裝
3. 設定帳號

```
git config --global user.name "user_name"
git config --global user.email "user_mail"
```
4. [建立公私鑰](https://docs.github.com/en/github/authenticating-to-github/connecting-to-github-with-ssh/generating-a-new-ssh-key-and-adding-it-to-the-ssh-agent)

```
ssh-keygen -t ed25519 -C "your_email@example.com"
# 詢問你公私鑰要存在哪個地方，預設是 /c/Users/you/.ssh/id_ed25519，你也可以自己改路徑，或是直接按 Enter 用預設的
Enter a file in which to save the key (/c/Users/you/.ssh/id_ed25519):[Press enter]
Enter passphrase (empty for no passphrase): [Type a passphrase] # 直接 Enter 即可
Enter same passphrase again: [Type passphrase again] # 直接 Enter 即可

clip < ~/.ssh/id_ed25519.pub # 複製該檔案的內容 -- 不會有回傳結果
(如果 clip 沒有效果，可以改成 type ~/.ssh/id_ed25519.pub，然後複製起來)
```

5. [添加到 github 上](https://docs.github.com/en/github/authenticating-to-github/connecting-to-github-with-ssh/adding-a-new-ssh-key-to-your-github-account)
